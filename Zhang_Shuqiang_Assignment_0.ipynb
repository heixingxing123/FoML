(c) Find vector $\mathbf{u}$ that is perpendicular to vector $\mathbf{v}$ above.

**Answer 1(c)**

If two vectors are perpendicular to each other, their dot product will be 0.

$$\mathbf{u} * \mathbf{v} = 0$$

$$\mathbf{u} * (1, 1) = 0$$

(d) Given point $A$ and vector $\mathbf{u}$ perpendicular to $\mathbf{v}$, provide a vector equation that every point $P$ on $\mathbf{u}$ must satisfy.
Reduce this equation to the form $ay_1 + by_2 = 1$, i.e. calculate $a$ and $b$ in terms of $A$ and $\mathbf{u}$.
$y_1, y_2$ are the coordinates of $P$. \\

*Hint: the inner product of vector $\mathbf{u}$ and a vector parallel to the line is zero.*

**Answer 1(d)**

Let P = $(y_1, y2)$, which is on the vector u.

Let vector AP represents the vector from A to P

$AP = P - A$

$AP * u = 0$

(P - A) * u = 0

$(y_1 - 1, y_2 - 2) * u = 0$

$(y_1 - 1, y_2 - 2) * (1, -1) = 0$ or $(y_1 - 1, y_2 - 2) * (-1, 1) = 0$

$y_1 - y_2 = -1$ or $-y_1 + y_2 = 1$

for $-y_1 + y_2 = 1$

(f) Consider a plane in three dimensions defined by a point $A$ and a normal vector $\mathbf{u}$, i.e. perpendicular to the plane. Given a point $B$ not on the plane, find the projection $B'$ of point $B$ onto the plane.

*Hint: You can make use of two properties about projection $B'$: $A$ being parallel to the plane and $B$ being parallel to vector $\mathbf{u}$.*

**Answer 1(f)**

Let:

$A = (A_x, A_y, A_z)$

$B = (B_x, B_y, B_z)$

$u = (u_x, u_y, u_z)$

---

Vector $AB = B - A$

We know the vector B'B is perpendicular to the plane, so vector B'B parallel to vector u. And we also know that A and B are on the plane.

The projection of vector AB onto vector u is $\frac{AB * u}{u * u}u$

Then we can infer that vector $B'B = \frac{AB * u}{u * u}u$


Given $B'B = B - B'$, so:

$B'B = \frac{AB * u}{u * u}u$

$B - B' = \frac{AB * u}{u * u}u$

$B' = B - \frac{AB * u}{u * u}u$

$B' = (B_x, B_y, B_z) - \frac{[(B_x, B_y, B_z) - (A_x, A_y, A_z)] * (u_x, u_y, u_z)}{(u_x, u_y, u_z) * (u_x, u_y, u_z)}(u_x, u_y, u_z)$

## 3. Sample Generation (5 pts)

(a) Write a function that uses the function `random_uniform_distribution(n)` (which generates data from a uniform distribution on `[0, 10)`) to generate samples which simulates the rolling of two six-sided dice. One is fair, meaning it has equal probability of landing on any side, and the other is loaded, meaning that it has the probability of landing on each side shown in the table below:

| Side        	| 1    	| 2    	| 3    	| 4    	| 5   	| 6    	|
|-------------	|------	|------	|------	|------	|-----	|------	|
| Probability 	| 0.17 	| 0.27 	| 0.17 	| 0.17 	| 0.05	| 0.17 	|

````python
###################### DO NOT EDIT ######################
def random_uniform_distribution(n):
    rng = np.random.default_rng()
    return 10 * rng.random(n)
````

````python
# your code here
def run_dice(n):
  fair_dice = []
  loaded_dice = []

  random_num = random_uniform_distribution(n)

  # fair die
  one_range = 1/6 * 10

  for i in range(n):
    if random_num[i] < one_range:
      fair_dice.append(1)
    elif random_num[i] < 2 * one_range:
      fair_dice.append(2)
    elif random_num[i] < 3 * one_range:
      fair_dice.append(3)
    elif random_num[i] < 4 * one_range:
      fair_dice.append(4)
    elif random_num[i] < 5 * one_range:
      fair_dice.append(5)
    elif random_num[i] < 6 * one_range:
      fair_dice.append(6)

  # loaded die
  for i in range(n):
    if random_num[i] < 0.17 * 10:
      loaded_dice.append(1)
    elif random_num[i] < 0.27 * 10 + 0.17 * 10:
      loaded_dice.append(2)
    elif random_num[i] < 0.17 * 10 + 0.27 * 10 + 0.17 * 10:
      loaded_dice.append(3)
    elif random_num[i] < 0.17 * 10 + 0.17 * 10 + 0.27 * 10 + 0.17 * 10:
      loaded_dice.append(4)
    elif random_num[i] < 0.05 * 10 + 0.17 * 10 + 0.17 * 10 + 0.27 * 10 + 0.17 * 10:
      loaded_dice.append(5)
    elif random_num[i] < 0.17 * 10 + 0.05 * 10 + 0.17 * 10 + 0.17 * 10 + 0.27 * 10 + 0.17 * 10:
      loaded_dice.append(6)

  return fair_dice, loaded_dice

n = 10000
fair_dice, loaded_dice = run_dice(n)

print("Check for fair dice")
print(fair_dice.count(1) / n)
print(fair_dice.count(2) / n)
print(fair_dice.count(3) / n)
print(fair_dice.count(4) / n)
print(fair_dice.count(5) / n)
print(fair_dice.count(6) / n)

print("Check for loaded dice")
print(loaded_dice.count(1) / n)
print(loaded_dice.count(2) / n)
print(loaded_dice.count(3) / n)
print(loaded_dice.count(4) / n)
print(loaded_dice.count(5) / n)
print(loaded_dice.count(6) / n)
````

a) Implement the k-nearest neighbours algorithm.

````python
# Your code here

import numpy as np

def knn_classifier(train, val, k):
  pred = []


  mean = np.mean(train, axis=0)
  std = np.std(train, axis=0)

  train_standardized = (train - mean) / std
  val_standardized = (val - mean) / std

  for i in range(len(val_standardized)):
        dist = np.sqrt(np.sum(np.square(val_standardized[i] - train_standardized), axis=1))
        idx = np.argsort(dist)
        top_k = idx[:k]
        pred.append(np.mean(train[top_k], axis=0))

  return pred
````

b) Split the dataset you created in Question 4 into a train and validation set. Plot the training and validation loss for different values of k, and use those plots to state which k value is optimal.

````python
# Your code here

np.random.shuffle(data)

data_normalized = (data - np.mean(data, axis=0)) / np.std(data, axis=0)

train_size = int(0.8 * len(data_normalized))
train_data = data_normalized[:train_size]
val_data = data_normalized[train_size:]

loss = []
for k in range(1, 10):
    pred = knn_classifier(train_data, val_data, k)
    loss.append(np.mean(np.square(pred - val_data)))

plt.plot(range(1, 10), loss)
plt.show()
````
